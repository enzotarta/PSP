```python

def readl(f):
    trainloss = []
    trainreg = []
    testloss = []
    testreg = []
    err = []
    c = -1
    for l in f:
        l = l.split(' ')
        if l[0] == 'LOG':
            trainloss.append([])
            trainreg.append([])
            testloss.append([])
            testreg.append([])
            err.append([])
            c+=1
        elif 'Epoch' in l[0]:
            trainloss[c].append(float(l[3]))
            trainreg[c].append(float(l[5]))
            testloss[c].append(float(l[8]))
            testreg[c].append(float(l[10]))
            err[c].append(float(l[12]))
    return trainloss,trainreg,testloss,testreg,err
        
```
